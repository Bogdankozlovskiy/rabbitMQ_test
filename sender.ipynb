```python
#sender
import pika
from datetime import datetime as dt

def main():
    credentials = pika.PlainCredentials('rabbitmq', 'rabbitmq')# login password
    parameters = pika.ConnectionParameters('rabbitmq', 5672, '/', credentials)
    connection = pika.BlockingConnection(parameters)
    
    channel = connection.channel()
    channel.queue_declare(queue='queuel')
    
    while True:
        t = dt.now().strftime('%T')
        channel.basic_publish(exchange='', routing_key='queuel', body='hello world: ' + t)
        print("[x] Send 'Hello World' " + t)
    
    connection.close()
main()
```

```python
import pika
import sys

credentials = pika.PlainCredentials('rabbitmq', 'rabbitmq')# login password
parameters = pika.ConnectionParameters('rabbitmq', 5672, '/', credentials)
connection = pika.BlockingConnection(parameters)
channel = connection.channel()

channel.queue_declare(queue='task_queue', durable=True)#сообщения в очереди долговечны

count = 0
while count < 100:
    message = "Hello.." + str(count)
    count += 1
    channel.basic_publish(
        exchange='',              # exchange используем по умолчанию, сообщения направляются в очередь 
        routing_key='task_queue', # с именем, указанным параметром routing_key
        body=message,
        properties=pika.BasicProperties(
            delivery_mode=2,  # сделать сообщения долговечными
        ))
    print(" [x] Sent %r" % message)

connection.close()
```

```python
import pika
import sys

credentials = pika.PlainCredentials('rabbitmq', 'rabbitmq')# login password
parameters = pika.ConnectionParameters('rabbitmq', 5672, '/', credentials)
connection = pika.BlockingConnection(parameters)
channel = connection.channel()

channel.exchange_declare(exchange='direct_logs', exchange_type='direct')

severities = 'info', 'warning', 'error'

count = 0
while count < 100:
    for severity in severities:
        message = 'Hello World! ' + str(count)
        channel.basic_publish(exchange='direct_logs', routing_key=severity, body=message)
        count += 1
        print(" [x] Sent %r:%r" % (severity, message))
    
connection.close()
```
